```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Settings
pd.options.display.float_format = '{:,.0f}'.format

colors = sns.color_palette('pastel')[0:10]
```

The dataset that I have chosen will be regarding Esports earnings. I have always been up to date on the gaming industry. However, I have only played games for fun and never had an exposure on how much who people earn going to gaming tournaments. Different countries should vary depending on the popularity of Esports but these are things that I am curious to find out.

I am interested in knowing the following:
* The amounts earned per video game and genre
* The difference of earning vs amount of players in each country
* Trend of number of tournaments and how much they earn
* Earning trend from games based on release year

```python
# importing dataset
df = pd.read_csv('Esport_Earnings.csv')
df.head()
```

```python
# Grouping by Genre to get sum of total earning by Genre
df_1 = df.groupby('Genre').sum().reset_index()
df_1 = df_1[['Genre', 'TotalMoney']]
df_1 = df_1.sort_values(by='TotalMoney', ascending=True)
df_1
```

```python
# Converting dataframe to thousands for dollar amounts only
df_1['TotalMoney'] = df_1['TotalMoney'] / 1000
df_1
```

```python
# Creating a horizantal barchart to show the earnings for each Genre
plt.barh(df_1['Genre'], df_1['TotalMoney'])
plt.title('Total Earnings Per Genre')
plt.ylabel('Genre')
plt.xlabel("Earnings in 000's")
plt.show()
```

```python
# formatting data to group amounts by country in order to see totals of players vs earnings
# and converting to thousands
# will divide Top_Country_Earnings over PlayerNo to calculate total earnings per player
# will calculate percentage of Country Top Earnings to total earnings
# Taking top 15
df_2 = df.groupby('Top_Country').sum().reset_index()
df_2 = df_2[['Top_Country', 'Top_Country_Earnings', 'TotalMoney', 'PlayerNo', 'TournamentNo']]
df_2 = df_2.astype({'PlayerNo' : 'float64', 'TournamentNo' : 'float64'})
df_2['Earning_Per_Player'] = df_2['Top_Country_Earnings'] / df_2['PlayerNo']
df_2['Pct_Top_Total'] = (df_2['Top_Country_Earnings'] - df_2['TotalMoney']) / df_2['TotalMoney']
df_2['Top_Country_Earnings'] = df_2['Top_Country_Earnings'] / 1000
df_2 = df_2.sort_values('Top_Country_Earnings', ascending=False)
df_2
```

```python
plt.scatter('Top_Country_Earnings', 'TotalMoney', c='Earning_Per_Player', s='Pct_Top_Total', data=df_2.head(15))
plt.show()
```

```python
# df to show number of tournaments and total earnings
df_3 = df.groupby('GameName').sum().reset_index()
df_3 = df_3[['GameName', 'TournamentNo', 'TotalMoney']]
df_3['TotalMoney'] = df_3['TotalMoney']
df_3
```

```python
plt.subplot(221)
plt.plot(df_3['TournamentNo'], df_3['TotalMoney'])
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
```

```python
# Grouping df by game release date
df_4 = df.groupby('Releaseyear').sum().reset_index()
df_4['Top_Country_Earnings'] = df_4['Top_Country_Earnings'] / 1000
df_4 = df_4.drop(df_4.index[0])
df_4
```

```python
plt.plot(df_4['Releaseyear'].tail(15), df_4['Top_Country_Earnings'].tail(15))
plt.plot(df_4['Releaseyear'].tail(15), df_4['TotalMoney'].tail(15))
plt.plot(df_4['Releaseyear'].tail(15), df_4['PlayerNo'].tail(15))
plt.plot(df_4['Releaseyear'].tail(15), df_4['TournamentNo'].tail(15))
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.show()
```
